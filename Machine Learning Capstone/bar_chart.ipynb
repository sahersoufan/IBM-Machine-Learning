```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "Type": ["KNN", "NMF", "NN", "NNR", "NNC"],
    "RMSE": [0.955, 1.061, 0.246, 0.1260, 0.106]
})

data.head()
```

```python
ax = sns.barplot(data, x="Type", y="RMSE", palette="viridis")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")
plt.title("RMSE vs Type of Collaborative Filtering")
plt.show()
```
